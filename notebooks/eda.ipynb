# Exploratory Data Analysis — Predictor de Resultados de Fútbol

Este notebook explora los datasets históricos de partidos de fútbol.

**Objetivos:**
- Entender la distribución de resultados (H/D/A)
- Analizar correlaciones entre features y resultado
- Identificar valores nulos y outliers
- Validar la calidad de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

sns.set_theme(style='whitegrid')
%matplotlib inline
```

```python
# Cargar todos los CSVs
files = glob.glob('../data/raw/*.csv')
print(f'Archivos encontrados: {len(files)}')

frames = [pd.read_csv(f, encoding='utf-8', on_bad_lines='skip') for f in files]
df = pd.concat(frames, ignore_index=True)
print(f'Total registros: {len(df)}')
df.head()
```

```python
# Info general y nulos
print(df.info())
print('\nNulos por columna:')
print(df.isnull().sum())
```

```python
# Distribución de resultados
fig, ax = plt.subplots(figsize=(8, 5))
df['FTR'].value_counts().plot(kind='bar', ax=ax, color=['#2ecc71', '#f39c12', '#e74c3c'])
ax.set_title('Distribución de Resultados')
ax.set_xlabel('Resultado (H=Local, D=Empate, A=Visitante)')
ax.set_ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# Estadísticas descriptivas de features numéricas
numeric_cols = ['FTHG', 'FTAG', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC']
available = [c for c in numeric_cols if c in df.columns]
df[available].describe()
```

```python
# Matriz de correlación
fig, ax = plt.subplots(figsize=(10, 8))
corr = df[available].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, ax=ax)
ax.set_title('Correlación entre Features')
plt.tight_layout()
plt.show()
```
